```python
# Importamos todas las librerías necesarias
import pandas as pd
import numpy as np
```

```python
# Leemos el dataset
df = pd.read_csv('MaribelDataset15-21.csv')
df.head()
```

```python
# Función para determinar OK o FAIL
def determinar_estado(row):
    if row['Milestone'] > 50:
        return row['Problem'] + ' OK'
    else:
        return row['Problem'] + ' FAIL'

# Creamos los identificadores Pn OK o Pn FAIL
# Aplicar la función a cada fila del DataFrame y crear la nueva columna
df['Estado'] = df.apply(lambda row: determinar_estado(row), axis=1)
```

```python
df.head()
```

```python
labels = df['Estado'].unique().tolist()
print(labels)
```

```python
# Eliminamos las columnas con valores NaN
df=df.dropna(axis=1)

# Ordenamos los valores según las sesiones
df.sort_values(by="Session")

# Creamos los identificadores problema-milestone
df['Problem'] = df['Problem'].replace({'P':''}, regex=True)
df['Compuesto'] = 'PROBLEM' + df.Problem.apply(str) + '-' + df.Milestone.apply(str)
df.head()
```

```python
for c in df.columns:
    print(c, df[c].dtype)
```

```python
# Convertir la columna 'sTime' en un objeto datetime
df['sTime'] = pd.to_datetime(df['sTime'])
df.head()
```

```python
# Calcular la diferencia entre el sTime menor y el sTime mayor de cada año
result = df.groupby('Year')['sTime'].apply(lambda x: (max(x) - min(x)).seconds)
print(result)
```

```python
df2015 = df[(df.Year == 'Y2015')]
# Dividir los valores de sTime en cuartiles
df2015['quartiles'] = pd.qcut(df2015['sTime'], q=4)
```

```python
print(df2015['quartiles'].unique().tolist())
```

```python
df = df[(df.Year == 'Y2016')]
# Dividir los valores de sTime en cuartiles
df['quartiles'] = pd.qcut(df['sTime'], q=4)
```

```python
quartiles = df['quartiles'].unique().tolist()
print(quartiles)
```
