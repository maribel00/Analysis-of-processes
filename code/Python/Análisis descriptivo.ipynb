```python
df['sTime'] = pd.to_datetime(df['sTime'], format='%d/%m/%Y %H:%M:%S')

# Agregar la columna del día de la semana
df['DayOfWeek'] = df['sTime'].dt.day_name()

# Agregar la columna de la hora del día
df['Hour'] = df['sTime'].dt.strftime('%H')

# Mostramos el dataframe con sus dos nuevas columnas
df.head()
```

```python
import matplotlib.pyplot as plt

# Datos de entrada
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
frequencies = [12830, 11500, 13432, 14372, 13297, 10310, 11289]

# Crear el histograma
plt.bar(weekdays, frequencies, color='paleturquoise', edgecolor='black')

# Configurar el título y etiquetas de los ejes
plt.title('Histogram of DayOfWeek', weight='bold')
plt.xlabel('DayOfWeek', weight='bold')
plt.ylabel('Frequency', weight='bold')

# Girar los días de la semana y ajustar el estilo de las etiquetas
plt.xticks(rotation=90, ha='right')

# Agregar rejilla de fondo
plt.grid(True, axis='y', linestyle='--')

# Mostrar el histograma
plt.tight_layout()
plt.show()
```

```python
# Obtener el índice de la fila con la fecha más tardía en cada grupo y sesión
indices = df.groupby(['Grupo', 'Session'])['sTime'].idxmax()

# Seleccionar las filas correspondientes a los índices encontrados
resultado = df.loc[indices].reset_index(drop=True)
```

```python
resultado.head()
```

```python
# Crear la columna 'Outcome' utilizando apply y una función lambda
resultado['Outcome'] = resultado['Milestone'].apply(lambda x: 'fail' if x < 100 else 'solved')
resultado.head()
```

```python
resultado.tail()
```

```python
# Agrupar por año, grupo y outcome y realizar conteo de sesiones
grouped = resultado.groupby(['Year', 'Grupo', 'Outcome']).size().reset_index(name='Count')

# Pivotear la tabla para obtener una columna por cada outcome
pivot = grouped.pivot_table(index=['Year', 'Grupo'], columns='Outcome', values='Count', fill_value=0).reset_index()

# Calcular la columna 'all' como suma de las columnas 'fail' y 'solved'
pivot['all'] = pivot['fail'] + pivot['solved']

# Mostrar el nuevo DataFrame resultante
pivot.head()
```

```python
# Obtener el código LaTeX de la tabla correspondiente al DataFrame
tabla_latex = pivot.to_latex(index=False)
print(tabla_latex)
```

```python
# Exportar el DataFrame a un archivo CSV
pivot.to_csv('MBSessionsExt.csv', index=False)
```
